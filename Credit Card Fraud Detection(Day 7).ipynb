```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("creditcard.csv")
df.head(5)

#class--0 represents normal transaction class--1 represents fraud transactions
```

```python
df.tail(5)
#the time shows 172792 s which is almmost 2 days of transaction
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Class'].value_counts()
#this is an unbalanced dataset as about 90% of the data points are in normal transactions
```

```python
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
df.groupby('Class').mean()
```

```python
# Building sample dataset for similaar distribution of normal transactions and fraudulent transactions

#we take 492 random datas from normal transactions to make the legal and fraud equal
legit_sample = legit.sample(n=492)
print(legit_sample)
```

```python
#concatenating the 2 dataframes

ndf = pd.concat([legit_sample,fraud],axis=0)
ndf.head(5)
```

```python
ndf['Class'].value_counts()
```

```python
ndf.groupby('Class').mean()
```

```python
X = ndf.drop(columns='Class',axis=1)
Y = ndf['Class']
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state = 2,test_size=0.2,stratify=Y)
#stratify is used to make sure that Y values are distributed equally among train and test
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction = model.predict(X_train)
train_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print(train_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print(test_data_accuracy)
```
